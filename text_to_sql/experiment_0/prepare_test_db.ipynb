```python
import pandas as pd
data = pd.read_csv('../../data/clean/catastici.csv')
```

```python
questions = pd.read_csv('../data/test_db_no_rent_raw.csv', sep=';')
questions.shape
```

# Level 0 QA

```python
questions_0 = questions[questions.level==0]
```

```python
seed = 42
data_0 = data.sample(25, random_state=seed)
```

```python
question_formatted = []
for (_,q),(_,d) in zip(questions_0.iterrows(), data_0.iterrows()):
    for key, value in d.items():
        if key in q['question']:
            placeholder = f"['{key}']"
            q['question'] = q['question'].replace(placeholder, f'"{value}"')
    question_formatted.append(q['question'])
questions_0['question_formatted'] = question_formatted
```

```python
question_formatted = []
for (_,q),(_,d) in zip(questions_1.iterrows(), data_1.iterrows()):
    for key, value in d.items():
        if key in q['question']:
            placeholder = f"['{key}']"
            q['question'] = q['question'].replace(placeholder, f'"{value}"')
    question_formatted.append(q['question'])
questions_1['question_formatted'] = question_formatted
```
